```python
# Implementasi Fungsi f(x) = 3x - 7 untuk -5 < x < 5, x âˆˆ Z
# a) Metode Konvensional
print("Metode Konvensional:")
for x in range(-4, 5):
    print(f"f({x}) = {3*x - 7}")
```

```python
# b) Menggunakan def
def f(x):
    return 3*x - 7

print("Menggunakan def:")
for x in range(-4, 5):
    print(f"f({x}) = {f(x)}")
```

```python
# c) Dengan List (anggota domain)
domain = list(range(-4, 5))
hasil = [3*x - 7 for x in domain]

print("Domain:", domain)
print("Hasil fungsi:", hasil)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# a) f(x) = 2x | -100 < x < 100
x = np.arange(-99, 100)
y = 2 * x

plt.figure()
plt.plot(x, y, label="f(x) = 2x", color='blue')
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.title("Grafik f(x) = 2x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
